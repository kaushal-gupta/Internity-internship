1) Write a Python programming to display a bar chart of the popularity of Programming Languages.
Sample data: 
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
import numpy as np
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]

y=[22.2, 17.6, 8.8, 8, 7.7, 6.7]


labels=['Java', 'Python','PHP', 'JavaScript',' C#',' C++']

plt.bar(x,y,width=0.7,tick_label=labels,color=['b','g','y','r','m'])

plt.xlabel("programing languages")
plt.ylabel("popularity")
plt.title("Popularity of programing languages")
plt.show()
```

2) Write a Python program to create a bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)

```python
import numpy as np
import matplotlib.pyplot as plt 

#hieght of men

mens=[22, 30, 35, 35, 26]


barwidth=0.4
#height of women

women=[25, 32, 30, 35, 29]




# The x position of bars
r1 = np.arange(len(mens))


# Create mens bars
plt.bar(r1-0.2, mens, width = barwidth, color = 'blue', edgecolor = 'black', label='mens')
plt.bar(r1+0.2, women, width = barwidth, color = 'r', edgecolor = 'black',label='womens')
plt.legend()



```

4) Write a Python programming to create a pie chart with a title of the popularity of Programming Languages.
Sample data: 
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7


```python
import numpy as np
import matplotlib.pyplot as plt

languages= ['Java',' Python',' PHP',' JavaScript',' C#',' C++']
Popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]

color=['r','g','b','m','y','black']

plt.pie(Popularity,labels=languages,colors=color,shadow=True)
plt.legend()
plt.show()
```
