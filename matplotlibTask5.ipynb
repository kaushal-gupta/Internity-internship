```python
import numpy as np
import matplotlib.pyplot as plt

t=np.arange(-360,360,.1)

y=np.cos(t*np.pi/180)


plt.plot(t,y,'--')
plt.grid()
plt.xlabel("x (in degrees)")
plt.ylabel("y")
plt.title(" Cosine graph")

plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt  
   
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(-360, 360, 0.1) 
y_sin = np.sin(x*np.pi/180) 
y_cos = np.cos(x*np.pi/180)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin) 
plt.title('Sine')  
plt.grid()
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos) 
plt.title('Cosine')  
plt.grid()
   
# Show the figure. 
plt.show()
```

```python

import matplotlib.pyplot as plt

x = [1,2,3,4,5]

y = [10,24,36,40,5]

label = ['One','Two','Three','Four','Five']

plt.bar(x,height=y,width = 0.7,tick_label=label,color=['green','blue','yellow','red'])

plt.xlabel('x-axis')

plt.ylabel('y-axis')

plt.title('Bar Graph')

plt.show()
```

```python
import matplotlib.pyplot as plt

marks = [2,50,70,40,30,45,50,45,43,40,44,60,
       7,13,51,18,90,77,32,21,20,40]

range = (0,100)
bins = 10

plt.hist(marks,bins,range,color='green',histtype='bar',rwidth=0.7)

plt.xlabel('marks')
plt.ylabel('Students')

plt.title('Histogram Plot')

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,10,3,45,42,33,4,4,53,53,34,55,30,4]
y = [23,20,8,6,34,20,30,67,3,33,55,33,22,3,4]
z = [43,60,78,68,4,50,50,6,3,36,25,3,22,3,4]

plt.scatter(x,y,color="green",label="stars",marker="*")
plt.scatter(x,z,color="blue",label="triangle",marker="v")

plt.legend()

plt.show()
```
